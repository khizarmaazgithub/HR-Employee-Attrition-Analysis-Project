Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)
```

 Load and explore the data

```python
df = pd.read_csv("C:/Users/KHIZAR MAAZ/Desktop/Dataset of Hr/WA_Fn-UseC_-HR-Employee-Attrition.csv")

print("Shape:",df.shape)
print("\nColumn names:")

print(df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

Basic data info

```python
print("Data Types:")
print(df.dtypes)

print("\nMissing values:")
print(df.isnull().sum())

print("\nStatistical Summary:")
df.describe()
```

 Check attrition distribution

```python
print("Attrition Count:")
print(df['Attrition'].value_counts())

print("\nAttrition Percentage:")
print(df['Attrition'].value_counts(normalize=True) * 100)
```

Create numeric attrition column

```python
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

overall_rate = df['Attrition_Flag'].mean() * 100
print(f"Overall Attrition Rate: {overall_rate:.1f}%")
```

 Drop useless columns

```python
cols_to_drop = ['EmployeeCount', 'Over18', 'StandardHours']
df = df.drop(columns=cols_to_drop, errors='ignore')

print('Remaining columns:', df.shape[1])
```

Chart 1: Attrition distribution (pie chart)

```python
colors = ['#4CAF50', '#F44336']
df['Attrition'].value_counts().plot(
    kind='pie', autopct='%1.1f%%',
    colors=colors, startangle=90
)
plt.title('Overall Employee Attrition Distribution')
plt.ylabel('')
plt.savefig('chart1_attrition_pie.png', bbox_inches='tight', dpi=150)
plt.show()
```

Chart 2: Attrition by Department

```python
dept = df.groupby('Department')['Attrition_Flag'].mean() * 100
dept = dept.sort_values(ascending=False).reset_index()
dept.columns = ['Department', 'Attrition_Rate']

plt.figure(figsize=(8, 4))
sns.barplot(data=dept, x='Department', y='Attrition_Rate', palette='Blues_r')
plt.title('Attrition Rate by Department (%)')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
for i, v in enumerate(dept['Attrition_Rate']):
    plt.text(i, v + 0.3, f'{v:.1f}%', ha='center', fontsize=10)
plt.savefig('chart2_dept_attrition.png', bbox_inches='tight', dpi=150)
plt.show()

print(dept)
```

Chart 3: Overtime vs Attrition

```python
overtime = df.groupby('OverTime')['Attrition_Flag'].mean() * 100
overtime = overtime.reset_index()
overtime.columns = ['OverTime', 'Attrition_Rate']

plt.figure(figsize=(6, 4))
sns.barplot(data=overtime, x='OverTime', y='Attrition_Rate',
            palette=['#4CAF50', '#F44336'])
plt.title('Attrition Rate: Overtime vs No Overtime (%)')
plt.ylabel('Attrition Rate (%)')
for i, v in enumerate(overtime['Attrition_Rate']):
    plt.text(i, v + 0.3, f'{v:.1f}%', ha='center', fontsize=11)
plt.savefig('chart3_overtime.png', bbox_inches='tight', dpi=150)
plt.show()

print(overtime)
```

Chart 4: Monthly Income distribution

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome',
            palette={'Yes': '#F44336', 'No': '#4CAF50'})
plt.title('Monthly Income: Employees who Left vs Stayed')
plt.xlabel('Attrition (Yes = Left, No = Stayed)')
plt.ylabel('Monthly Income ($)')
plt.savefig('chart4_income.png', bbox_inches='tight', dpi=150)
plt.show()

print("Avg Income - Left:", df[df['Attrition']=='Yes']['MonthlyIncome'].mean().round(0))
print("Avg Income - Stayed:", df[df['Attrition']=='No']['MonthlyIncome'].mean().round(0))
```

Chart 5: Job Satisfaction vs Attrition

```python
sat = df.groupby('JobSatisfaction')['Attrition_Flag'].mean() * 100
sat = sat.reset_index()
sat.columns = ['JobSatisfaction', 'Attrition_Rate']
sat['Label'] = sat['JobSatisfaction'].map({
    1:'1-Low', 2:'2-Medium', 3:'3-High', 4:'4-Very High'
})

plt.figure(figsize=(8, 4))
sns.barplot(data=sat, x='Label', y='Attrition_Rate', palette='RdYlGn')
plt.title('Attrition Rate by Job Satisfaction Level (%)')
plt.ylabel('Attrition Rate (%)')
for i, v in enumerate(sat['Attrition_Rate']):
    plt.text(i, v + 0.3, f'{v:.1f}%', ha='center', fontsize=10)
plt.savefig('chart5_satisfaction.png', bbox_inches='tight', dpi=150)
plt.show()
```

Chart 6: Age distribution

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Age', hue='Attrition', bins=20,
             palette={'Yes': '#F44336', 'No': '#4CAF50'}, alpha=0.7)
plt.title('Age Distribution: Attrition vs Retained')
plt.xlabel('Age')
plt.savefig('chart6_age.png', bbox_inches='tight', dpi=150)
plt.show()
```

Chart 7: Years at Company vs Attrition

```python
tenure = df.groupby('YearsAtCompany')['Attrition_Flag'].mean() * 100

plt.figure(figsize=(12, 4))
tenure.plot(kind='bar', color='steelblue', alpha=0.8)
plt.title('Attrition Rate by Years at Company (%)')
plt.xlabel('Years at Company')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=45)
plt.savefig('chart7_tenure.png', bbox_inches='tight', dpi=150)
plt.show()
```

Chart 8: Correlation Heatmap

```python
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(16, 12))
corr = numeric_df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, cmap='coolwarm', annot=False,
            linewidths=0.3, fmt='.2f')
plt.title('Correlation Heatmap — All Numerical Features')
plt.savefig('chart8_heatmap.png', bbox_inches='tight', dpi=150)
plt.show()
```

Key findings summary

```python
print("=" * 50)
print("KEY FINDINGS FROM EDA")
print("=" * 50)

print(f"\n1. Overall attrition rate: {df['Attrition_Flag'].mean()*100:.1f}%")

dept_top = df.groupby('Department')['Attrition_Flag'].mean() * 100
print(f"\n2. Highest attrition dept: {dept_top.idxmax()} ({dept_top.max():.1f}%)")

ot_yes = df[df['OverTime']=='Yes']['Attrition_Flag'].mean() * 100
ot_no = df[df['OverTime']=='No']['Attrition_Flag'].mean() * 100
print(f"\n3. Attrition with OverTime: {ot_yes:.1f}% vs without: {ot_no:.1f}%")

low_sat = df[df['JobSatisfaction']==1]['Attrition_Flag'].mean() * 100
high_sat = df[df['JobSatisfaction']==4]['Attrition_Flag'].mean() * 100
print(f"\n4. Low satisfaction attrition: {low_sat:.1f}% vs High: {high_sat:.1f}%")

inc_left = df[df['Attrition']=='Yes']['MonthlyIncome'].mean()
inc_stay = df[df['Attrition']=='No']['MonthlyIncome'].mean()
print(f"\n5. Avg income left: ${inc_left:.0f} vs stayed: ${inc_stay:.0f}")
```
